```python
#Importing Libraries 
from pyspark.sql import SparkSession
import pandas as pd 
```

```python

# Starting a sparksession 
spark = SparkSession.builder\
                    .master("local")\
                    .appName('Firstprogram')\
                    .getOrCreate()
sc=spark.sparkContext



```

```python
# Read the input file and Calculating words count
text_file = sc.textFile("updatedFile.txt")
# non_bg_lined = text_file.filter(lambda x: x.startsWith("BG:"))
# counts = text_file.flatMap(lambda line: line.split(" ")) \
#                             .map(lambda word: (word, 1)) \
#                            .reduceByKey(lambda x, y: x + y)
```

```python
counts = text_file.flatMap(lambda line: line.split(" ")) \
                            .map(lambda word: (word, 1)) \
                           .reduceByKey(lambda x, y: x + y)
```

```python
output = counts.collect()

data_df = list()
for (word, count) in output:
  data_df.append((word, count))
```

```python
import pandas as pd 
df = pd.DataFrame(data_df, columns=['word', 'count'])

print(df)
```

```python
df.to_csv("word_count.csv")
```

```python
# Stopping Spark-Session and Spark context
sc.stop()
spark.stop()
```

```python
# Update the given values to the new columns \

data.loc[data['Title'] == 'Forest Gump', 'views'] = 10000000
data.loc[data['Title'] == 'The Usual Suspects', 'views'] = 7500000
data.loc[data['Title'] == 'Rear Window', 'views'] = 6000000
data.loc[data['Title'] == 'North by Northwest', 'views'] = 4000000
data.loc[data['Title'] == 'The Secret in Their Eyes', 'views'] = 3000000
data.loc[data['Title'] == 'Spotlight', 'views'] = 1000000
```
